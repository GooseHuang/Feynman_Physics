```python
import numpy as np
from matplotlib import pyplot as plt
import math


class Planet():
    def __init__(self,x,y,vx,vy,t,M=1,G=1,m=1):
        self.x = x
        self.y = y
        self.vx = vx
        self.vy = vy
        self.t = t
        self.M = M
        self.G = G
        self.m = m
        self.ax,self.ay = self.calculate_accelation()
        self.records ={'x': [self.x],
                       'y': [self.y],
                       'vx': [self.vx],
                       'vy': [self.vy],
                       'ax':[self.ax],
                       'ay': [self.ay],
                       't': [self.t]}
        
        
    def calculate_accelation(self):
        r = ((self.x)**2 + (self.y)**2)**0.5
        f = -1/r**2 * self.G * self.M * self.m
        fx = self.x/r * f
        fy = self.y/r * f
        ax = fx/self.m
        ay = fy/self.m
        return ax,ay
        
    def move(self, delta):
        self.x += self.vx * delta
        self.y += self.vy * delta
        self.vx += self.ax * delta
        self.vy += self.ay * delta
        self.ax, self.ay = self.calculate_accelation()
        self.t += delta
        self.recording()

        
    def info(self):
        print('x:',self.x)
        print('y:',self.y)
        print('vx:',self.vx)
        print('vy:',self.vy)
        print('ax:',self.ax)
        print('ay:',self.ay)
        print('t:',self.t)
        print()
        
    def recording(self):
        self.records['x'].append(self.x)
        self.records['y'].append(self.y)
        self.records['vx'].append(self.vx)
        self.records['vy'].append(self.vy)
        self.records['ax'].append(self.ax)
        self.records['ay'].append(self.ay)
        self.records['t'].append(self.t)
        
    def plot_x(self,show=True):
        x = np.array(self.records['t'])/self.w
        y = np.array(self.records['x'])
        plt.plot(x,y,'r')
        plt.title('X')
        if show:
            plt.show() 
        
    def plot_v(self,show=True):
        x = np.array(self.records['t'])/self.w
        y = np.array(self.records['v'])
        plt.plot(x, y, 'g')
        plt.title('V')
        if show:
            plt.show() 
        
    def plot_a(self,show=True):
        x = np.array(self.records['t'])/self.w
        y = np.array(self.records['a'])
        plt.plot(x,y,'b')
        plt.title('a')
        if show:
            plt.show() 
        
    def plot(self, show=True):
        self.plot_x(show)
        self.plot_v(show)
        self.plot_a(show)
        plt.show()
        
dt = 0.000002
x = 0.5
y = 0
vx = 0 
vy = -0.5
t = 0

planet =Planet(x,y,vx,vy,t,M=1,G=1,m=1)
for i in range(500000):
#     planet.info()
    planet.move(dt)
# spring.plot()
```
