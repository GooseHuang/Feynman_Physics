```python
import numpy as np
from matplotlib import pyplot as plt
import math


class Spring():
    def __init__(self,x=-1,v=0,t=0,w=math.pi):
        self.x = x
        self.v = v
        self.a = x
        self.w = w
        self.t = t
        self.records = {'x': [self.x],
                       'v': [self.v],
                       'a': [self.a],
                       't': [self.t]}
        
        
#     def move(self, delta):
#         self.x += self.v * delta
#         self.v += -self.a * delta
#         self.a = self.x
#         self.t += delta
#         self.recording()

    def move(self, delta):
        v0 = self.v
        self.v += -self.a * delta
        self.x += (self.v + v0)/2 * delta
        self.a = self.x
        self.t += delta
        self.recording()
        
        
    def info(self):
        print('x:',self.x)
        print('v:',self.v)
        print('a:',self.a)
        print('t:',self.t)
        print()
        
    def recording(self):
        self.records['x'].append(self.x)
        self.records['v'].append(self.v)
        self.records['a'].append(self.a)
        self.records['t'].append(self.t)
        
    def plot_x(self,show=True):
        x = np.array(self.records['t'])/self.w
        y = np.array(self.records['x'])
        plt.plot(x,y,'r')
        plt.title('X')
        if show:
            plt.show() 
        
    def plot_v(self,show=True):
        x = np.array(self.records['t'])/self.w
        y = np.array(self.records['v'])
        plt.plot(x, y, 'g')
        plt.title('V')
        if show:
            plt.show() 
        
    def plot_a(self,show=True):
        x = np.array(self.records['t'])/self.w
        y = np.array(self.records['a'])
        plt.plot(x,y,'b')
        plt.title('a')
        if show:
            plt.show() 
        
    def plot(self, show=True):
        self.plot_x(show)
        self.plot_v(show)
        self.plot_a(show)
        plt.show()
        
dt = 0.00001
spring = Spring()
for i in range(1000000):
#     spring.info()
    spring.move(dt)
    
spring.plot()
```
