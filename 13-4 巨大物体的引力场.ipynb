```python
import math
import numpy as np


class Earth():
    def __init__(self):
        self.base = (0,0)
        self.P = (20,0)
        self.R = 5
        self.step = 10000
        self.delta = math.pi / self.step
        self.mu = 1
        self.G = 1
        
        
    def calculate_force_field(self, theta):
        r = abs(self.R * math.sin(theta))
        s = 2 * math.pi * r * self.delta * self.R
        m = s * self.mu
        
        L = np.array((self.R*math.cos(theta) - self.P[0], self.R*math.sin(theta) - self.P[1]))
        
        force_field = (self.G * m / np.dot(L,L)**1.5)* L
        
#         print(round(theta/math.pi,2), L)
        
        return force_field
        
```

```python
earth = Earth()
for i in range(10000):
    force_field = earth.calculate_force_field(i *  1 * math.pi / earth.step)
    print(round(i *  1 * math.pi / earth.step, 2), force_field)
```
